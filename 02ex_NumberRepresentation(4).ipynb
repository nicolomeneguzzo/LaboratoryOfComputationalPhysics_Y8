2\. Write a function that converts a 32 bit word into a single precision floating point (i.e. interprets the various bits as sign, mantissa and exponent)

```python
def float(x):
    x = str(x)
    res = (-1) ** int(x[0]) * 2 ** (int(x[1:9], 2) - 127) * (1 + sum([int(x[i+8]) * 2 ** (-i) for i in range(1, 24)]))
    return res
```

**Tips**: define a new variable by adding a smaller and smaller value (proceeding similarly to prob. 2) to an original variable and check the point where the two are the same 

```python
def machine_epsilon():
    eps = 1.0
    while 1.0 + eps != 1.0:
        eps /= 2.0
    # Quando esce dal ciclo, eps è troppo piccolo, quindi moltiplichiamo per 2
    return eps * 2

eps = machine_epsilon()
print("Machine epsilon:", eps)
```

5\. Write a function that takes in input three parameters $a$, $b$ and $c$ and prints out the two solutions to the quadratic equation $ax^2+bx+c=0$ using the standard formula:
$$
x=\frac{-b\pm\sqrt{b^2-4ac}}{2a}
$$

(a) use the program to compute the solution for $a=0.001$, $b=1000$ and $c=0.001$

(b) re-express the standard solution formula by multiplying top and bottom by $-b\mp\sqrt{b^2-4ac}$ and again find the solution for $a=0.001$, $b=1000$ and $c=0.001$. How does it compare with what previously obtained? Why?

(c) write a function that compute the roots of a quadratic equation accurately in all cases

```python
import math

def quadratic_standard(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant < 0:
        print("No real roots")
        return None, None
    sqrt_discriminant = math.sqrt(discriminant)
    x1 = (-b + sqrt_discriminant) / (2*a)
    x2 = (-b - sqrt_discriminant) / (2*a)
    print(f"Standard formula roots: x1 = {x1}, x2 = {x2}")
    return x1, x2
```

```python
def quadratic_stable(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant < 0:
        print("No real roots")
        return None, None
    sqrt_discriminant = math.sqrt(discriminant)
    
    if b >= 0:
        x1 = (-2 * c) / (b + sqrt_discriminant)
        x2 = (-b - sqrt_discriminant) / (2 * a)
    else:
        x1 = (-b + sqrt_discriminant) / (2 * a)
        x2 = (-2 * c) / (b - sqrt_discriminant)
    
    print(f"Stable formula roots: x1 = {x1}, x2 = {x2}")
    return x1, x2
```

```python
def quadratic_roots(a, b, c):
    if a == 0:
        if b == 0:
            print("No solution")
            return None, None
        return -c/b, None  # Linear case

    discriminant = b**2 - 4*a*c
    if discriminant < 0:
        print("No real roots")
        return None, None
    
    sqrt_discriminant = math.sqrt(discriminant)

    # Scelta stabile della formula
    if b >= 0:
        x1 = (-b - sqrt_discriminant) / (2*a)
        x2 = (2*c) / (-b - sqrt_discriminant)
    else:
        x1 = (-b + sqrt_discriminant) / (2*a)
        x2 = (2*c) / (-b + sqrt_discriminant)

    return x1, x2
```

6\. Write a program that implements the function $f(x)=x(x−1)$

(a) Calculate the derivative of the function at the point $x = 1$ using the derivative definition:

$$
\frac{{\rm d}f}{{\rm d}x} = \lim_{\delta\to0} \frac{f(x+\delta)-f(x)}{\delta}
$$

with $\delta = 10^{−2}$. Calculate the true value of the same derivative analytically and compare with the answer your program gives. The two will not agree perfectly. Why not?

(b) Repeat the calculation for $\delta = 10^{−4}, 10^{−6}, 10^{−8}, 10^{−10}, 10^{−12}$ and $10^{−14}$. How does the accuracy scales with $\delta$?

```python
def f(x):
    return x * (x - 1)

def derivative_numerical(x, delta):
    return (f(x + delta) - f(x)) / delta

x0 = 1.0
delta = 1e-2

numerical_derivative = derivative_numerical(x0, delta)
analytical_derivative = 2*x0 - 1

print("Numerical derivative (delta=1e-2):", numerical_derivative)
print("Analytical derivative:", analytical_derivative)
print("Difference:", numerical_derivative - analytical_derivative)
```

```python
deltas = [1e-2, 1e-4, 1e-6, 1e-8, 1e-10, 1e-12, 1e-14]

for delta in deltas:
    num_deriv = derivative_numerical(x0, delta)
    diff = num_deriv - analytical_derivative
    print(f"delta={delta:.0e}, numerical={num_deriv:.16f}, diff={diff:.2e}")
```

7\. Consider the integral of the semicircle of radius 1:
$$
I=\int_{-1}^{1} \sqrt(1-x^2) {\rm d}x
$$
which it's known to be $I=\frac{\pi}{2}=1.57079632679...$.
Alternatively we can use the Riemann definition of the integral:
$$
I=\lim_{N\to\infty} \sum_{k=1}^{N} h y_k 
$$

with $h=2/N$ the width of each of the $N$ slices the domain is divided into and where
$y_k$ is the value of the function at the $k-$th slice.

(a) Write a programe to compute the integral with $N=100$. How does the result compares to the true value?

(b) How much can $N$ be increased if the computation needs to be run in less than a second? What is the gain in running it for 1 minute? 


```python
import math

def semicircle(x):
    return math.sqrt(1 - x**2)

def riemann_integral(N):
    h = 2 / N
    total = 0.0
    for k in range(N):
        xk = -1 + (k + 0.5) * h  # midpoint
        total += semicircle(xk) * h
    return total

N = 100
I_approx = riemann_integral(N)
I_true = math.pi / 2

print(f"N={N}, Approximation={I_approx:.10f}, True value={I_true:.10f}, Error={abs(I_approx-I_true):.2e}")
```

```python
import time

N_test = [10**4, 10**5, 10**6, 10**7]
for N in N_test:
    start = time.time()
    riemann_integral(N)
    elapsed = time.time() - start
    print(f"N={N}, Time={elapsed:.4f} s")
```
