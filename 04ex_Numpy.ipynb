2\. Find the outer product of the following two vecotrs

```python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])
```

Do this in the following ways:

   * Using the function outer in numpy
   * Using a nested for loop or list comprehension
   * Using numpy broadcasting operatoins


````python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])

outer_product = np.outer(u, v)
print(f'prodotto esterno u⊗v= \n {outer_product}')

out_prod = np.array([[i*j for i in u] for j in v])
print(f'outer product= \n {out_prod}')

ut = u.reshape(4,1)
print(f'outer product= \n {ut * v}')
````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a raw and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
import numpy as np
import matplotlib.pyplot as plt
n_w = 1000
n_step = 200

steps = np.random.randint(0, 2, size=(n_w, n_step))
steps = 2 * steps -1

dist = np.cumsum(steps, axis=1)
square_dist = dist**2

squaredist_mean = np.mean(square_dist, axis=0)
dist_mean = np.sqrt(squaredist_mean)
# print(len(dist_mean), len(n_step))

time = np.arange(n_step)
# print(time)
# plt.plot(time, squaredist_mean, label='quadrato della distanza')
plt.plot(time, dist_mean, label='distanza media')
plt.xlabel('time')
plt.ylabel('mean dist')
plt.title('Diffusion via 1D random walk')
plt.legend()
plt.show()

#yes, the result match espectations. The mean squared displacement grows linearly with time,
#and the typical distance from the origin grows as the square root pf time, 
#which is the hallmark of normal diffusion
````

  * Check the content by looking within the file
  * Load the data (use an appropriate numpy method) into a 2D array
  * Create arrays out of the columns, the arrays being (in order): *year*, *hares*, *lynxes*, *carrots* 
  * Plot the 3 populations over the years
  * Compute the main statistical properties of the dataset (mean, std, correlations, etc.)
  * Which species has the highest population each year?

Do you feel there is some evident correlation here? [Studies](https://www.enr.gov.nt.ca/en/services/lynx/lynx-snowshoe-hare-cycle) tend to believe so.

````python

import matplotlib.pyplot as plt

data = np.loadtxt('populations.txt', skiprows=1)

#head populations.txt su bash
# print(data.shape)
# print(data[:5])

year = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]

plt.plot(year, hares, label='hares')
plt.plot(year, lynxes, label='lynxes')
plt.plot(year, carrots, label='carrots')
plt.xlabel('years')
plt.ylabel('populations')
plt.legend()
plt.show()

print(f'mean hares= {np.mean(hares):.7}, std= {np.std(hares):.7}')
print(f'mean lynxes= {np.mean(lynxes):.7}, std= {np.std(lynxes):.7}')
print(f'mean carrots= {np.mean(carrots):.7}, std= {np.std(carrots):.7}')

print('\nMin/Max:')
print(f'Hares: min={np.min(hares)}, max={np.max(hares)}')
print(f'Lynxes: min={np.min(lynxes)}, max={np.max(lynxes)}')
print(f'Carrots: min={np.min(carrots)}, max={np.max(carrots)}')

# matrice con le tre popolazioni
populations = np.array([hares, lynxes, carrots]).T
corr_matrix = np.corrcoef(populations)
print(f"\nCorrelation matrix=\n{corr_matrix}")

max_indices = np.argmax(populations, axis=1) #per ogni riga prende il valore più alto tra le colonne

species_names = ["Hares", "Lynxes", "Carrots"]
max_species_per_year = [species_names[i] for i in max_indices]

# visualizza primi 10 anni
for y, s in zip(year[:10], max_species_per_year[:10]):
    print(f"Year {int(y)}: {s} has the highest population")

#correlazione: evident correlations between the species.
#hares and lynxes:positive correlation, when hares increase, lynxes increase shortly after,
#hares and carrots:negative correlation, when hares increase carrots decreas
#lynxes and carrots: weak negative correlations
````
