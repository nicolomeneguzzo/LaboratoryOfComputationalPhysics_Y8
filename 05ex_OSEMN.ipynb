4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
with open('credit_card.dat', 'rb') as cred:
    lines = cred.readlines()

crco = [[chr(int(lines[i][j:j+6], 2)) for j in range(0, len(lines[i]), 6)][:-1] for i in range(len(lines))][:-1]
for i in range(len(crco)): print(''.join(crco[i]))
```

```python
#versione piu leggibile

with open('credit_card.dat', 'rb') as cred:
    lines = cred.readlines()

decoded = []

for line in lines:
    chars = []
    for j in range(0, len(line), 6):
        bits = line[j:j+6]
        if len(bits) < 6:
            continue
        chars.append(chr(int(bits, 2)))
    decoded.append(''.join(chars))

for card in decoded:
    print(card)
```
