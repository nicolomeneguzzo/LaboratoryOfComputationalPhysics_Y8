1\.d Find out which is the letter with most counts.

```python
max=0
maxl=""

for i in {A..Z}; do
	count=$(grep -c "^$i" LCP_22-23_students.csv)
	if (( count > max )); then
		max=$count
		maxl=$i
	fi
done
```

```python
awk -F, '
{ first = substr($1,1,1); count[first]++ }
END {
    max = 0
    for (c in count) {
        if (count[c] > max) {
            max = count[c]
            maxc = c
        }
    }
    print "La lettera con più righe è " maxc " con " max " righe"
}
' LCP_22-23_students.csv
```

1\.e Assume an obvious numbering of the students in the file (first line is 1, second line is 2, etc.), group students "modulo 18", i.e. 1,19,37,.. 2,20,38,.. etc. and put each group in a separate file  

```python
awk '{
    group = ((NR-1) % 18) + 1      # modulo 18
    filename = "group_" group ".csv"
    print > filename
}' LCP_22-23_students.csv
```
